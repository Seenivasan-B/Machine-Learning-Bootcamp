```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"E:\study materials\python\02 Machine Learning - Udemy\UNZIP_FOR_NOTEBOOKS_FINAL\DATA\cancer_tumor_data_features.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

```python
from sklearn.decomposition import PCA
```

```python
help(PCA)
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dictionary = load_breast_cancer()
```

```python
pca_model = PCA(n_components=2)
```

```python
pc_results = pca_model.fit_transform(scaled_X)
```

```python
plt.scatter(pc_results[:,0],pc_results[:,1])
```

```python
plt.scatter(pc_results[:,0],pc_results[:,1],c = cancer_dictionary['target'])
```

```python
pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_,index=['PC1','PC2'],columns = df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(20,3),dpi=200)
sns.heatmap(df_comp,annot=True)
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_30 = PCA(n_components=30)
```

```python
pca_30.fit(scaled_X)
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30),explained_variance)
plt.xlabel('Num of components')
plt.ylabel('Variance Explained')
```
