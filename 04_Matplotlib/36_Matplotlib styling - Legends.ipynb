```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
```

```python
x
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x, x**2, label = 'X vs X^2')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x, x**2, label = 'X vs X^2')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x, x**2, label = 'X vs X^2')

ax.legend(loc = 'lower left')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x, x**2, label = 'X vs X^2')

ax.legend(loc=4)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x, x**2, label = 'X vs X^2')

ax.legend(loc=(1.1,0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x, x**2, label = 'X vs X^2')

ax.legend(loc=(-0.5,0.5))
```
