```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
b = a ** 4
```

```python
a
```

```python
b
```

```python
fig = plt.figure(figsize=(10,3),dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig = plt.figure(figsize=(4,2),dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig('new_fig.png',bbox_inches='tight')
```
