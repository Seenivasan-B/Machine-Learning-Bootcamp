```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color = 'purple')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color = '#650dbd')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color = '#21cc29')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color = '#21cc29',label='x vs x')
ax.plot(x,x+1,color = 'purple',label='x vs x+1')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color = '#21cc29',lw=10)
ax.plot(x,x+1,color = 'purple',linewidth=5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color = '#21cc29',lw=10,ls='--')
ax.plot(x,x+1,color = 'purple',linewidth=5,linestyle='-.')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x+1,color = 'purple',linewidth=5)
lines[0].set_dashes([5,2,5,2])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x+1,color = 'purple',linewidth=5)
lines[0].set_dashes([5,2,10,2])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x+1,color = 'purple',linewidth=2,marker='o',ls='--')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color = 'red',lw=2,marker='+',ls='--',ms=10)
ax.plot(x,x+1,color = 'purple',linewidth=2,marker='o',ls='--',markersize=20)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color = 'red',lw=2,marker='s',ls='--',ms=10,markerfacecolor='red',markeredgewidth=4,markeredgecolor='orange')
```
