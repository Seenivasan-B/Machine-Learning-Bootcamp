```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
b = a ** 4
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y);
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b);
```

```python
fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

#small axes
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(x,y)

plt.show()
```

```python
fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#small axes
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show()
```
